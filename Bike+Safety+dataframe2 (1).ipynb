```python
import pandas as pd

def reorganize(df):
    df = df.set_index(['Unnamed: 0'])
    df = df.dropna()
    df = df.rename(columns = {'Unnamed: 1':'Total','Unnamed: 2':'Male', 'Unnamed: 3': 'Female', 'Unnamed: 4': 'Helmet', 
                      'Unnamed: 5':'No Helmet', 'Unnamed: 6':'Passenger', 'Unnamed: 7':'No Passenger', 
                      'Unnamed: 8':'On Sidewalk', 'Unnamed: 9':'On Road'})
    df['Intersection'] = intersection
    return df 
```

```python
xl = pd.ExcelFile('54546_Wilcocks St EB from Spadina Ave to Huron St.xls') 

nrows = xl.book.sheet_by_index(2).nrows

wilcocks_spadina_EB = xl.parse(2, skiprows = 5, skip_footer = nrows - 21).dropna(axis=0, how='all')

intersection = 'Wilcocks and Spadina EB'

reorganize(wilcocks_spadina_EB)
```
