```python
'''EGLINTON AND EDEN VALLEY 2015 COUNTS'''

import pandas as pd

#I deleted empty rows within the tables before uploading this file
# I changed 'west' and 'east' to 'WB' and 'EB' before uploading this file

'''load excel file for parsing'''
xl = pd.ExcelFile('Eglinton & Eden Valley.xls')
nrows = xl.book.sheet_by_index(0).nrows

'''parse tables into seperate dataframes to remove empty rows'''
df1 = xl.parse(0, skiprows = 3, skip_footer = nrows-33).dropna(axis=1, how='all')
df2 = xl.parse(0, skiprows = 42, skip_footer = nrows-72).dropna(axis=1, how='all')
df3 = xl.parse(0, skiprows = 81, skip_footer = nrows - 111 ).dropna(axis = 1, how = 'all')
df4 = xl.parse(0, skiprows = 119, skip_footer = nrows - 149 ).dropna(axis = 1, how = 'all')

'''reorganize df1'''
df1 = df1.set_index(['Unnamed: 1'])
df1 = df1.drop(['Unnamed: 6', 'Unnamed: 11'], axis = 1)
df1 = df1.T
df1 = df1.fillna(method = 'ffill')
df1 = df1.set_index(['Date:'])

'''reorganize df2'''
df2 = df2.set_index(['Unnamed: 1'])
df2 = df2.drop(['Unnamed: 6', 'Unnamed: 11'], axis = 1)
df2 = df2.T
df2 = df2.fillna(method = 'ffill')
df2 = df2.set_index(['Date:'])

'''reorganize df3'''
df3 = df3.set_index(['Unnamed: 1'])
df3 = df3.drop(['Unnamed: 6', 'Unnamed: 11'], axis = 1)
df3 = df3.T
df3 = df3.fillna(method = 'ffill')
df3 = df3.set_index(['Date:'])

'''reorganize df4'''
df4 = df4.set_index(['Unnamed: 1'])
df4 = df4.T
df4 = df4.fillna(method = 'ffill')
df4 = df4.set_index(['Date:'])

'''recombine into 1 dataframe'''
eglinton_edenvalley2015_df = pd.concat([df1, df2, df3, df4])

'''add intersection column'''
eglinton_edenvalley2015_df['Intersection'] = 'Eglinton and Eden Valley'

'''rename columns/index for consistency'''
eglinton_edenvalley2015_df = eglinton_edenvalley2015_df.rename(columns = {'Day:':'Day','Weather:':'Weather','Hour':'Direction'})
eglinton_edenvalley2015_df.index.name = 'Date'

'''check final result'''
eglinton_edenvalley2015_df
```

```python
'''ADELAIDE AND SPADINA 2015 COUNTS'''

import pandas as pd

#It was noted in the spreadsheet that this is a one-way eastbound bike lane

xl = pd.ExcelFile('Adelaide & Spadina 2015.xlsx')
nrows = xl.book.sheet_by_index(0).nrows

df1 = xl.parse(0, skiprows = 3, skip_footer = nrows-32).dropna(axis=1, how='all')
df2 = xl.parse(0, skiprows = 36, skip_footer = nrows-65).dropna(axis=1, how='all')

df1 = df1.set_index(['Unnamed: 0'])
df1 = df1.T
df1 = df1.set_index(['Date'])

df2 = df2.set_index(['Unnamed: 0'])
df2 = df2.T
df2 = df2.set_index(['Date'])

adelaide_spadina2015_df = pd.concat([df1, df2])

adelaide_spadina2015_df['Direction'] = 'EB'
adelaide_spadina2015_df['Intersection'] = 'Adelaide and Spadina'

adelaide_spadina2015_df = adelaide_spadina2015_df.rename(columns = {'Total':'Totals'})

adelaide_spadina2015_df
```

```python
'''RICHMOND AND SPADINA 2015 COUNTS'''

import pandas as pd

#It was noted in the spreadsheet that this is a one-way westbound bike lane

xl = pd.ExcelFile('Richmond & Spadina 2015.xlsx')
nrows = xl.book.sheet_by_index(0).nrows

df1 = xl.parse(0, skiprows = 3, skip_footer = nrows-32).dropna(axis=1, how='all')
df2 = xl.parse(0, skiprows = 36, skip_footer = nrows-65).dropna(axis=1, how='all')

df1 = df1.set_index(['Unnamed: 0'])
df1 = df1.T
df1 = df1.set_index(['Date'])

df2 = df2.set_index(['Unnamed: 0'])
df2 = df2.T
df2 = df2.set_index(['Date'])

richmond_spadina2015_df = pd.concat([df1, df2])

richmond_spadina2015_df['Direction'] = 'WB'
richmond_spadina2015_df['Intersection'] = 'Richmond and Spadina'

richmond_spadina2015_df = richmond_spadina2015_df.rename(columns = {'Total':'Totals'})

richmond_spadina2015_df
```

```python
'''SIMCOE AND ADELAIDE 2015 COUNTS'''

import pandas as pd

#note that I deleted empty rows within the tables before uploading this file 

xl = pd.ExcelFile('Simcoe & Adelaide.xls')
nrows = xl.book.sheet_by_index(0).nrows

df1 = xl.parse(0, skiprows = 3, skip_footer = nrows-33).dropna(axis=1, how='all')
df2 = xl.parse(0, skiprows=42, skip_footer = nrows - 72).dropna(axis=1, how='all')
df3 = xl.parse(0, skiprows = 81, skip_footer = nrows-111).dropna(axis=1, how='all')

df1 = df1.set_index(['Unnamed: 1'])
df1 = df1.drop(['Unnamed: 6', 'Unnamed: 11'], axis = 1)
df1 = df1.T
df1 = df1.fillna(method = 'ffill')
df1 = df1.set_index(['Date:'])

df2 = df2.set_index(['Unnamed: 1'])
df2 = df2.drop(['Unnamed: 6', 'Unnamed: 11'], axis = 1)
df2 = df2.T
df2 = df2.fillna(method = 'ffill')
df2 = df2.set_index(['Date:'])

df3 = df3.set_index(['Unnamed: 1'])
df3 = df3.drop(['Unnamed: 6', 'Unnamed: 11'], axis = 1)
df3 = df3.T
df3 = df3.fillna(method = 'ffill')
df3 = df3.set_index(['Date:'])

simcoe_adelaide2015_df = pd.concat([df1, df2, df3])

simcoe_adelaide2015_df['Intersection'] = 'Simcoe and Adelaide'
simcoe_adelaide2015_df = simcoe_adelaide2015_df.rename(columns={'Day:':'Day', 'Weather:':'Weather', 'Hour':'Direction'})
simcoe_adelaide2015_df.index.name = 'Date'


simcoe_adelaide2015_df
```

```python
'''WELLESLEY AND PARLIAMENT 2015 COUNTS'''

import pandas as pd

#prior to importing the file, I removed empty rows in the tables

xl = pd.ExcelFile('Wellesley & Parliament.xls')
nrows = xl.book.sheet_by_index(0).nrows

df1 = xl.parse(0, skiprows = 2, skip_footer = nrows-33).dropna(axis=1, how='all')
df2 = xl.parse(0, skiprows=36, skip_footer = nrows-67).dropna(axis=1, how='all')
df3 = xl.parse(0, skiprows = 69, skip_footer = nrows - 100 ).dropna(axis = 1, how = 'all')

df1 = df1.set_index(['Unnamed: 1'])
df1 = df1.drop(['Unnamed: 5', 'Unnamed: 9'], axis = 1)
df1 = df1.T
df1 = df1.fillna(method = 'ffill')
df1 = df1.set_index(['Date:'])

df2 = df2.set_index(['Unnamed: 1'])
df2 = df2.drop(['Unnamed: 5', 'Unnamed: 9'], axis = 1)
df2 = df2.T
df2 = df2.fillna(method = 'ffill')
df2 = df2.set_index(['Date:'])

df3 = df3.set_index(['Unnamed: 1'])
df3 = df3.drop(['Unnamed: 5', 'Unnamed: 9'], axis = 1)
df3 = df3.T
df3 = df3.fillna(method = 'ffill')
df3 = df3.set_index(['Date:'])

wellesley_parliament2015_df = pd.concat([df1, df2, df3]) 

wellesley_parliament2015_df['Intersection'] = 'Wellesley and Parliament'
wellesley_parliament2015_df = wellesley_parliament2015_df.drop(['Total'], axis = 1)
wellesley_parliament2015_df = wellesley_parliament2015_df.rename(columns={'Day:':'Day', 'Weather:':'Weather', 'Hour':'Direction',
                                                                         'Subtotal': 'Totals'})
wellesley_parliament2015_df.index.name = 'Date'

wellesley_parliament2015_df
```
