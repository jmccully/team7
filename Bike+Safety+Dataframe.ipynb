```python
import pandas as pd

'''Load spreadsheet and select table'''

xl = pd.ExcelFile('54514_Sussex Ave WB from Spadina Ave to Huron St.xls') 

nrows = xl.book.sheet_by_index(2).nrows


sussex_spadina_WB = xl.parse(2, skiprows = 5, skip_footer = nrows - 21).dropna(axis=0, how='all')
```

```python
sussex_spadina_WB = sussex_spadina_WB.set_index(['Unnamed: 0'])

sussex_spadina_WB = sussex_spadina_WB.dropna()

sussex_spadina_WB = sussex_spadina_WB.rename(columns = {'Unnamed: 1':'Total','Unnamed: 2':'Male', 'Unnamed: 3': 'Female', 'Unnamed: 4': 'Helmet', 
                      'Unnamed: 5':'No Helmet', 'Unnamed: 6':'Passenger', 'Unnamed: 7':'No Passenger', 
                      'Unnamed: 8':'On Sidewalk', 'Unnamed: 9':'On Road'})

sussex_spadina_WB['Intersection'] = 'Sussex and Spadina WB'
```
