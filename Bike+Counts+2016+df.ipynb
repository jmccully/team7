```python
'''ADELAIDE AND JARVIS 2016 COUNTS'''

import pandas as pd

# it is noted in the spreadsheet that direction is one-way (EB)
# combined "temp" and "prec" into a new "weather" column before uploading the file (to maintain consistency with the other spreadsheets)

'''load excel file for data parsing'''
xl = pd.ExcelFile('Adelaide & Jarvis.xlsx')
nrows = xl.book.sheet_by_index(0).nrows

'''parse tables in excel file to exclude empty rows'''
df1 = xl.parse(0, skiprows = 2, skip_footer = nrows - 31).dropna(axis=1, how='all')
df2 = xl.parse(0, skiprows = 34, skip_footer = nrows - 63).dropna(axis=1, how='all')

'''reorganize df1 to desirable columns/indexes'''
df1 = df1.set_index(['Unnamed: 0'])
df1 = df1.T
df1 = df1.set_index(['Date'])

'''reorganize df2 to desirable columns/indexs'''
df2 = df2.set_index(['Unnamed: 0'])
df2 = df2.T
df2 = df2.set_index(['Date'])


'''combine resulting dataframes'''
adelaide_jarvis2016_df = pd.concat([df1, df2])


'''add intersection and direction columns, streamline other column names'''
adelaide_jarvis2016_df['Intersection'] = 'Adelaide and Jarvis'
adelaide_jarvis2016_df['Direction'] = 'EB'
adelaide_jarvis2016_df = adelaide_jarvis2016_df.rename(columns={'Total':'Totals'})

'''check final dataframe'''
adelaide_jarvis2016_df
```

```python
'''ADELAIDE AND SPADINA 2016 COUNTS'''

import pandas as pd

# it is noted in the spreadsheet that direction is one-way (EB)

xl = pd.ExcelFile('Adelaide & Spadina.xlsx')
nrows = xl.book.sheet_by_index(0).nrows

df1 = xl.parse(0, skiprows = 3, skip_footer = nrows - 32).dropna(axis=1, how='all')
df2 = xl.parse(0, skiprows = 36, skip_footer = nrows - 65).dropna(axis=1, how='all')

df1 = df1.set_index(['Unnamed: 0'])
df1 = df1.T
df1 = df1.set_index(['Date'])

df2 = df2.set_index(['Unnamed: 0'])
df2 = df2.T
df2 = df2.set_index(['Date'])

adelaide_spadina2016_df = pd.concat([df1, df2])

adelaide_spadina2016_df['Intersection'] = 'Adelaide and Spadina'
adelaide_spadina2016_df['Direction'] = 'EB'
adelaide_spadina2016_df = adelaide_spadina2016_df.rename(columns={'Total':'Totals'})

adelaide_spadina2016_df
```

```python
'''DUNDAS AND CARLAW 2016 COUNTS'''

import pandas as pd

#I deleted empty rows within table before uploading this file
#I will take "subtotal" as the "totals" column - skip Total from the spreadsheet

xl = pd.ExcelFile('Dundas & Carlaw.xls')
nrows = xl.book.sheet_by_index(0).nrows

df1 = xl.parse(0, skiprows = 2, skip_footer = nrows-32).dropna(axis=1, how='all')
df2 = xl.parse(0, skiprows = 36, skip_footer = nrows-66).dropna(axis=1, how='all')
df3 = xl.parse(0, skiprows = 69, skip_footer = nrows - 99 ).dropna(axis = 1, how = 'all')

df1 = df1.set_index(['Unnamed: 1'])
df1 = df1.drop(['Unnamed: 5', 'Unnamed: 9'], axis = 1)
df1 = df1.T
df1 = df1.fillna(method = 'ffill')
df1 = df1.set_index(['Date:'])

df2 = df2.set_index(['Unnamed: 1'])
df2 = df2.drop(['Unnamed: 5', 'Unnamed: 9'], axis = 1)
df2 = df2.T
df2 = df2.fillna(method = 'ffill')
df2 = df2.set_index(['Date:'])

df3 = df3.set_index(['Unnamed: 1'])
df3 = df3.T
df3 = df3.fillna(method = 'ffill')
df3 = df3.set_index(['Date:'])

dundas_carlaw2016_df = pd.concat([df1, df2, df3])

dundas_carlaw2016_df['Intersection'] = 'Dundas and Carlaw'
dundas_carlaw2016_df = dundas_carlaw2016_df.rename(columns = {'Day:':'Day', 'Weather:':'Weather','Hour':'Direction','Subtotal':'Totals'})
dundas_carlaw2016_df.index.name = 'Date'

dundas_carlaw2016_df
```
